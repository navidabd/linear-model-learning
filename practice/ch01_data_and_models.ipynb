```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "study hours": [1, 2, 3, 4, 5,6,7,8],
    "exam score": [50, 60, 70, 80, 90,100,110,120]
})

print(df)
```

```python
plt.scatter(df["study hours"], df["exam score"])

plt.xlabel("study hours")
plt.ylabel("exam score")
plt.title("Study Hours vs Exam Score")

plt.show()
```

```python
X=np.array(df["study hours"])
y=np.array(df["exam score"])

print("X:", X)
print("y:", y)
```

```python
print("average study hours:", np.mean(X))
print("average exam score:", np.mean(y))
```

```python
correlation= np.corrcoef(X, y)[0, 1]
print("correlation between study hours and exam score:", correlation)
```
