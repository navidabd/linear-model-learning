```python
import pandas as pd
import statsmodels.api as sm

#data
df=pd.DataFrame({
    "education":[12,14,16,18,20],
    "experience":[1,2,4,6,8],
    "salary":[30,35,50,65,80]
})

#define x
X = df[["education","experience"]]

#add intercept
X = sm.add_constant(X)

#define y
Y = df["salary"]

mode=sm.OLS(Y,X).fit()
print(mode.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

X = df[["education","experience"]]

vif = pd.DataFrame()
vif["Variable"] = X.columns

vif["VIF"] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

print(vif)
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

#data
df=pd.DataFrame({
    "education":[12,14,16,18,20],
    "experience":[1,2,4,6,8],
    "salary":[30,35,50,65,80]
})

#define x
X = df[["education","experience"]]

#add intercept
X = sm.add_constant(X)

#define y
Y = df["salary"]

mode=sm.OLS(Y,X).fit()
print(mode.summary())
vif = pd.DataFrame()
vif["Variable"] = X.columns

vif["VIF"] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

print(vif)
```
