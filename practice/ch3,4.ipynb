```python
import numpy as np
from sklearn.linear_model import LinearRegression

x= np.array([
    [16,5],
    [18,8],
    [12,2],
    [20,10]

])

y = np.array([70,100,50,120])

model = LinearRegression()

model.fit(x,y)

print(model.coef_)
print(model.intercept_)
```

```python
import statsmodels.api as sm
import numpy as np

X = np.array([
    [16,5],
    [18,8],
    [12,2],
    [20,10]
])

y = np.array([70,100,50,120])

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```
